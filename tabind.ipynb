```python
import itertools
```

```python
# Get all combinations of letters in tabind such as a, na, da, etc.
letters = 'tabind'
combinations = []

for i in range(1, len(letters) + 1):
    for perm in itertools.permutations(letters, i):
        combinations.append(''.join(perm))
```

```python
with open("Collins Scrabble Words (2019).txt", "r") as file:
    wordList = file.read().splitlines()

newWordList = [word.lower() for word in wordList if 1 <= len(word) <= 6]
legalWords = set()

for word in newWordList:
    for index, letter in enumerate(word):
        if letter not in 'tabind':
            break
        if index == len(word) - 1:
            legalWords.add(word)

legalWords = sorted(legalWords)
```

```python
all_words = [word for word in combinations if word in legalWords]

print(all_words)
print(len(all_words))
```
